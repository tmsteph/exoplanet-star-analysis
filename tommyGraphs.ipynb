```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
planets = pd.read_csv("planets_with_habitability.csv")
planets = planets.dropna()
sns.set(style="ticks")
```

```python
habitable = planets[planets.habitable != 0] 

habitable
```

```python
st_teff	st_tefferr1	st_tefferr2	st_tefflim	st_rad	st_raderr1	st_raderr2	st_radlim	st_mass	st_masserr1	st_masserr2	
```

```python
f, ax = plt.subplots(figsize=(7, 6))

sns.scatterplot(x="st_mass", y="habitable", data=planets) 


ax.xaxis.grid(False)
ax.set(xlabel="Star Mass")
ax.set(ylabel="Habitability")

sns.despine(trim=True, left=True)
```

```python
f, ax = plt.subplots(figsize=(7, 6))

sns.scatterplot(x="st_rad", y="habitable", data=planets) 


ax.xaxis.grid(False)
ax.set(xlabel="Star Radius")
ax.set(ylabel="Habitability")

sns.despine(trim=True, left=True)
```

```python
f, ax = plt.subplots(figsize=(7, 6))

sns.scatterplot(x="st_teff", y="habitable", data=planets) 


ax.xaxis.grid(False)
ax.set(xlabel="Star Temp")
ax.set(ylabel="Habitability")

sns.despine(trim=True, left=True)
```

```python
plt.figure(figsize=(12, 12))
ax = sns.scatterplot(x='st_mass', y='st_teff',
                     hue='habitable', size='st_rad',
                     sizes=(40, 400), alpha=.7,
                     data=planets)
# limiting x and y axis's values
ax.axis([0,2, 2000, 8000])
ax.set_xlabel('Stellar Mass (Solar Mass)', fontsize=13)
ax.set_ylabel('Effective Temperature (Kelvin)', fontsize=13)
ax.set_title('Temperature - Mass Relation with Habitability', fontsize=18)
```
